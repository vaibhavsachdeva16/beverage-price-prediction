```python
bins = [18, 25, 35, 45, 55, 70, 120]

labels = [
    '18-25',
    '26-35',
    '36-45',
    '46-55',
    '56-70',
    '70+'
]

survey_df['age_group'] = pd.cut(survey_df['age'], bins=bins, labels=labels, right=True, include_lowest=True)

survey_df[['age', 'age_group']]
```

```python
# remove original age column

survey_df = survey_df.drop('age', axis=1)

survey_df
```

### 2. Create 'cf_ab_score' (Consume Frequency and Awareness Brand Score)

**Objective**: Create a new column `cf_ab_score` by combining the information from `consume_frequency(weekly)` and `awareness_of_other_brands` columns.

* Use the following mappings for `consume_frequency(weekly)`:
    * '0-2 times' -> 1
    * '3-4 times' -> 2
    * '5-7 times' -> 3
* Use the following mappings for `awareness_of_other_brands`:
    * '0 to 1' -> 1
    * '2 to 4' -> 2
    * 'above 4' -> 3
    
* Calculate cf_ab_score using the following formula:

$$ cf__ab__score = frequency score / (awareness score + frequency score) $$

* Round the result to two decimal places.

You need to convert the two categorical columns into numerical scores first, then compute the new feature cf_ab_score using the given formula.

#### Map consume_frequency(weekly) to Scores

```python
frequency_map = {
    '0-2 times': 1,
    '3-4 times': 2,
    '5-7 times': 3
}

survey_df['frequency_score'] = survey_df['consume_frequency(weekly)'].map(frequency_map)

survey_df
```

#### Map awareness_of_other_brands to Scores

```python
awareness_map = {
    '0 to 1': 1,
    '2 to 4': 2,
    'above 4': 3
}

survey_df['awareness_score'] = survey_df['awareness_of_other_brands'].map(awareness_map)

survey_df
```

#### Calculate cf_ab_score

```python
survey_df['cf_ab_score'] = round(survey_df['frequency_score'] / (
    survey_df['awareness_score'] + survey_df['frequency_score']
), 2)

survey_df
```

### 3. Create Zone Affluence Score (ZAS)

**Objective**: Calculate the `zas_score` using a combination of the `zone` and `income_levels` columns.

* Use the following mappings for the `zone` column:
    * 'Urban' -> 3
    * 'Metro' -> 4
    * 'Rural' -> 1
    * 'Semi-Urban' -> 2

* Use the following mappings for the `income_levels` column:
    * '<10L' -> 1
    * '10L - 15L' -> 2
    * '16L - 25L' -> 3
    * '26L - 35L' -> 4
    * '> 35L' -> 5
    * 'Not Reported' -> 0

* Calculate the `zas_score` using the following formula:
$$ zas__score = zone score * income score $$

```python
zone_map = {
    'Urban': 3,
    'Metro': 4,
    'Rural': 1,
    'Semi-Urban': 2
}

survey_df['zone_score'] = survey_df['zone'].map(zone_map)
```

```python
income_map = {
    '<10L': 1,
    '10L - 15L': 2,
    '16L - 25L': 3,
    '26L - 35L': 4,
    '> 35L': 5,
    'Not Reported': 0
}

survey_df['income_score'] = survey_df['income_levels'].map(income_map)
```

```python
survey_df['zas_score'] = survey_df['zone_score'] * survey_df['income_score']
```

## Predictive Modeling

### 1. Prepare Features and Target Variables:

* Create a feature matrix 'X' and target variable 'y'. Ensure that 'X' does not include columns such as respondent_id and the target variable (price_range)

```python
X = survey_df_final.drop(['respondent_id', 'price_range'], axis=1)

y = survey_df_final['price_range']
```

```python
y.unique()
```

### 2. Data Splitting:

* Split the dataset into training and test sets, using 75% of the data for training and 25% for testing. In train_test_split function call, please use random_state value of 42

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
X_train.shape, X_test.shape
```

### 3. Feature Encoding

* Apply appropriate encoding techniques to the features:
    * Use Label Encoding for the following columns: age_group, income_levels, health_concerns, consume_frequency(weekly), and preferable_consumption_size
    * Apply one-hot encoding to all remaining categorical columns.
    * Ensure the target variable (price_range) is also label encoded.

```python
# Target Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

# Save target encoder

pickle.dump(le, open("files/label_encoder.pkl", "wb"))
```

```python
# Define columns

ordinal_cols = [
    'age_group',
    'income_levels',
    'health_concerns',
    'consume_frequency(weekly)',
    'preferable_consumption_size'
]

ordinal_categories = [
    ['18-25', '26-35', '36-45', '46-55', '56-70', '70+'],  # age_group
    ['<10L', '10L - 15L', '16L - 25L', '26L - 35L', '> 35L', 'Not Reported'],  # income_levels (adjust if different)
    ['Low (Not very concerned)', 'Medium (Moderately health-conscious)', 'High (Very health-conscious)'],  # health_concerns
    # You MUST define correct order for below based on your data:
    sorted(survey_df_final['consume_frequency(weekly)'].unique()), # consume_frequency(weekly)
    ['Small (250 ml)', 'Medium (500 ml)', 'Large (1 L)'] # preferable_consumption_size
]

nominal_cols = [col for col in X.columns if col not in ordinal_cols]
```

```python
# Preprocessing

from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('ord', OrdinalEncoder(categories=ordinal_categories), ordinal_cols),
        ('nom', OneHotEncoder(handle_unknown='ignore', sparse_output=False), nominal_cols)
    ]
)
```

### 4. Model Selection:

* Test the following machine learning models on the prepared data:
    * Gaussian Naive Bayes
    * Logistic Regression
    * SVM
    * Random Forest
    * XGBoost
    * Light GBM

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

models = {
    "Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(),
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    "LightGBM": LGBMClassifier()
}
```

```python
# Train using Pipeline

from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

results = {}

for name, model in models.items():
    
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    pipeline.fit(X_train, y_train)
    
    y_pred = pipeline.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)
    
    results[name] = acc
    print(f"{name}: {acc:.4f}")
```

### 5. Performance Evaluation:

* For each model calculate and print the accuracy and the classification report.

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report

results = {}

for name, model in models.items():
    
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    pipeline.fit(X_train, y_train)
    
    y_pred = pipeline.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)
    
    print(f"\n===== {name} =====")
    print("Accuracy:", acc)
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    results[name] = acc
```

### 6. Model Comparison:

* Track the performance of each model and select the best-performing model for the next steps

```python
from sklearn.metrics import classification_report

best_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LGBMClassifier())
])

best_pipeline.fit(X_train, y_train)

y_pred = best_pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Save the Best Model

best_model_name = max(results, key=results.get)
print("\nBest Model:", best_model_name)

best_model = models[best_model_name]

final_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', best_model)
])

final_pipeline.fit(X_train, y_train)
```
